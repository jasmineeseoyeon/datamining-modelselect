#### 'required_car_parking_space', 'arrival_month', 'arrival_year', 'repeated_guest'를 category로

```python
categorical = ['type_of_meal_plan', 'required_car_parking_space', 'room_type_reserved', 'arrival_month', 'arrival_year',  'market_segment_type', 'repeated_guest','booking_status']
for i in categorical:
  hotel[i] = hotel[i].astype('category')
hotel.info()
```

```python
numeric = hotel.select_dtypes(include=['int64', 'float64']).columns
numeric
```

```python
# 정규화 (∵값의 범위가 큰 변수가 더 많은 영향을 미치기 때문)
from sklearn.preprocessing import MinMaxScaler
t_features = hotel[numeric]
scaler_m = MinMaxScaler()
t_features_m = pd.DataFrame(scaler_m.fit_transform(t_features))
t_features_m.columns = ['no_of_adults', 'no_of_children', 'no_of_weekend_nights', 'no_of_week_nights', 'lead_time','arrival_date', 'no_of_previous_cancellations', 'no_of_previous_bookings_not_canceled',  'avg_price_per_room', 'no_of_special_requests']
t_features_m.describe()
```

```python
categorical_data = hotel[['type_of_meal_plan', 'required_car_parking_space', 'room_type_reserved', 'arrival_year', 'arrival_month', 'market_segment_type', 'repeated_guest', 'booking_status']]

hotel = pd.concat([t_features_m, categorical_data], axis=1)
hotel
```

```python
# 'booking_status'를 숫자형으로 변환
hotel['booking_status_numeric'] = hotel['booking_status'].map({'Canceled': 1, 'Not_Canceled': 0})

# 수치형 데이터 선택 (booking_status_numeric을 포함)
numeric_columns = hotel.select_dtypes(include=[np.number]).copy()

# 'booking_status_numeric' 열을 numeric_columns에 추가
if 'booking_status_numeric' not in numeric_columns.columns:
    numeric_columns['booking_status_numeric'] = hotel['booking_status_numeric']

# 상관 행렬 계산
correlation_matrix = numeric_columns.corr()

# 'booking_status_numeric'과의 상관 관계
booking_status_corr = correlation_matrix['booking_status_numeric'].drop('booking_status_numeric')

# 히트맵 그리기
plt.figure(figsize=(10, 8))
sns.heatmap(booking_status_corr.to_frame(), annot=True, cmap='coolwarm', cbar=True, linewidths=0.5)
plt.title('Correlation of Attributes with Booking Status')
plt.xlabel('Booking Status')
plt.ylabel('Attributes')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score  # 정확도 계산 함수
from sklearn.model_selection import train_test_split, learning_curve
from imblearn.over_sampling import SMOTE

# 데이터 로드 및 전처리
def preprocess_data_with_smote(file_path, target_column):
    dataset = pd.read_csv(file_path)
    encoders = {}
    for column in dataset.columns:
        if dataset[column].dtype == 'object':
            from sklearn.preprocessing import LabelEncoder
            encoder = LabelEncoder()
            dataset[column] = encoder.fit_transform(dataset[column])
            encoders[column] = encoder

    X = dataset.drop(columns=[target_column])
    y = dataset[target_column]

    smote = SMOTE(random_state=42)
    X_resampled, y_resampled = smote.fit_resample(X, y)
    return X_resampled, y_resampled

# 학습 곡선 생성
def plot_learning_curve(model, X, y, title):
    train_sizes, train_scores, test_scores = learning_curve(model, X, y, cv=5, scoring='accuracy')
    train_scores_mean = np.mean(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)

    plt.figure(figsize=(10, 6))
    plt.plot(train_sizes, train_scores_mean, 'o-', label="Training Score")
    plt.plot(train_sizes, test_scores_mean, 'o-', label="Validation Score")
    plt.title(title)
    plt.xlabel("Training Set Size")
    plt.ylabel("Accuracy")
    plt.legend(loc="best")
    plt.grid()
    plt.show()

# 데이터 로드 및 분석 실행
file_path = 'Hotel Reservations_original.csv'  # 데이터 파일 경로
target_column = "booking_status"
X_resampled, y_resampled = preprocess_data_with_smote(file_path, target_column)

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled)

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

y_pred = rf_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)  # 정확도 계산
print(f"전체 데이터 사용 시 테스트 정확도: {accuracy:.2f}")

plot_learning_curve(rf_model, X_resampled, y_resampled, "Learning Curve (전체 데이터)")
```
